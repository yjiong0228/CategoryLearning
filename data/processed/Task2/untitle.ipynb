```python
import os

# 定义映射规则
mapping = {
    1: 125, 4: 126, 7: 127, 10: 128,
    13: 129, 16: 130, 19: 131, 22: 132,
    2: 225, 5: 226, 8: 227, 11: 228,
    14: 229, 17: 230, 20: 231, 23: 232,
    3: 325, 6: 326, 9: 327, 12: 328,
    15: 329, 18: 330, 21: 331, 24: 332
}

# 遍历当前文件夹下的所有文件
for filename in os.listdir('.'):
    # 只处理符合 Task2_*_convert_rec.csv 格式的文件
    if filename.startswith('Task2_') and filename.endswith('_convert_rec.csv'):
        # 提取中间的数字部分
        num_part = filename.split('_')[1]
        # 转换为整数
        try:
            old_num = int(num_part)
        except ValueError:
            continue
        # 查找新数字
        if old_num in mapping:
            new_num = mapping[old_num]
            new_filename = f'Task2_{new_num}_convert_rec.csv'
            # 重命名文件
            os.rename(filename, new_filename)
            print(f"已重命名: {filename} -> {new_filename}")
        else:
            print(f"跳过: {filename}（无对应映射）")
```
